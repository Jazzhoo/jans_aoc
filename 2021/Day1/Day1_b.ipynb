Reading the input file

```python
file = open('day1_input', 'r')
lines = file.readlines()
```

```python
no_increases = 0
no_decreases = 0
sums = []

for i in range(len(lines[:-2])):
    sum=0
    for j in range(3):
        sum+=int(lines[i+j])
    sums.append(sum)
    #print(sum)


for k in range(len(sums)):    
    if sums[k] > sums[k-1]:
        no_increases = no_increases + 1
        print(f"{int(sums[k])}: increase --> {no_increases}")

    else:
        print(f"{int(sums[k])}: decrease")
        no_decreases = no_decreases + 1
```
