```python
file = open('day1_input', 'r')
lines = file.readlines()
```

```python
no_increases = 0
no_decreases = 0

for i in range(len(lines)):
    if lines[i] > lines[i-1]:
        no_increases = no_increases + 1
        print(f"{int(lines[i])}: increase --> {no_increases}")
        
    else:
        print(f"{int(lines[i])}: decrease")
        no_decreases = no_decreases + 1

#no idea why I had to add 1 to the result. The right answer was 1681
```
